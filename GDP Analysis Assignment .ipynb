### Why is (Primary + Secondary + Tertiary) not equal to total GDP?

GVA (Gross value added) = (Primary + Secondary + Tertiary) 

GVA is the grand total of all revenues, from final sales and (net) subsidies, which are incomes into businesses. Those incomes are then used to cover expenses (wages & salaries, dividends), savings (profits, depreciation), and (indirect) taxes.

GVA is sector specific, and GDP is calculated by summation of GVA of all sectors of economy with taxes added and subsidies are deducted.

#### GVA + taxes on products - subsidies on products = Total GDP
This is why Total GDP in not equal to (Primary + Secondary + Tertiary)

```python
c1 = Sectors_df['PerCapitaGSDP_Percentile_rank']>=0.85
c1_states = Sectors_df[c1]
c1_states
```

```python
#calculating the sum of GSDP of C1 States
Sum_GDP_c1 = c1_states['Gross State Domestic Product'].sum()

#dropping unnecessary columns, i.e. sectors etc.
c1_states = c1_states.drop(['Primary', 'Secondary', 'Tertiary'], axis=1)

#drop columns from GSDP to PerCapitaGSDP_Percentile_rank
c1_states = c1_states.drop(c1_states.iloc[:, 12:], axis = 1)

#set State as index 
c1_states = c1_states.set_index('State')

#changing the type of column from object to float
c1_states = c1_states.astype(float)

#Calculating sum of all sub-sectors for C1 states
c1_states = c1_states.append(c1_states.agg(['sum']))

#apply transpose
c1_states = c1_states.T

#add new column % contribution to GDP of differenct sub-sectors
c1_states['% contribution to GDP'] = (c1_states['sum']/Sum_GDP_c1)*100

#Round off the columns
c1_states = c1_states.round(2)

c1_states
```
